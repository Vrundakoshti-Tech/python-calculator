```python
# This functions adds two numbers
def add(x, y):
    return x + y

# This function subtracts two numbers
def subtract(x, y):
    return x - y

# This function multiplys two numbers
def multiply(x, y):
    return x * y

# This function divides two numbers
def divide(x, y):
    return x / y

# This function is for exponentiation
def power(x, y):
    return x ** y

# This function is for square root
def square_root(x):
    if x < 0:
        return "Error: Cannot take square root of a negative number!!!"
    return x**0.5 # or math.sqrt(x) if you import math


print("Select operation.")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")
print("5. Expontiation")
print("6. Square root")

while True:
    # take input from the user
    choice = input("Enter choice(1/2/3/4/5/6): ")

    #  check if the choice is one of the six options
    if choice in ('1', '2', '3', '4', '5', '6'):
        try:
            if choice == '6': # For square root, only one number is needed
               num1 = float(input("Which number's square root do you want???: "))
               num2 = None # Or you can just skip asking for 2nd number entirely
            else: # For other operations, two numbers are needed
               num1 = float(input("Tell me the 1st number: "))
               num2 = float(input("Tell me the 2nd number: "))
        except ValueError:
            print("Invalid input. Please enter a valid number!!")
            continue

        if choice == '1':
            print(num1, "+",  num2, "=", add(num1, num2))

        elif choice == '2':
            print(num1, "-", num2, "=", subtract(num1, num2))

        elif choice == '3':
            print(num1, "*", num2, "=", multiply(num1, num2))

        elif choice == '4':
            print(num1, "/", num2, "=", divide(num1, num2))

        elif choice == '5':
            print(num1, "**", num2, "=", power(num1, num2))

        elif  choice == '6':
            print(f"square root of {num1} = {square_root(num1)}" )

        # check if the user wants another calculation
        # break the while loop is amswer is no
        next_calculation = input("wanna do another calculation? (yes/no): ")
        if next_calculation == "no":
            break
    else:
        print("Invalid input")
```
